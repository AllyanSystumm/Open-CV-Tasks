```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow  # Import the correct function for Colab

# Read the image
img = cv2.imread("/content/babar.jpeg")

img_gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
print(img_gray.shape)
# Display the original image
cv2_imshow(img)

# Display the grayscale image
cv2_imshow(img_gray)

# Flip the image horizontally
img_flipped_vertically= cv2.flip(img, 0)

# Display the vertically flipped image
cv2_imshow(img_flipped_vertically)
```
